```python
import numpy as np
import matplotlib.pyplot as plt
from utility_functions import fit_bass_model, predict_bass_model, compute_cumulative_adoption, compute_instantaneous_adoption, plot_data

# iRobot revenue data (2012-2023) in million USD
years = np.array([2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023])
revenue = np.array([436.24, 487.4, 556.85, 616.78, 660.6, 883.9, 1092.6, 1214, 1430.4, 1564.99, 1183.38, 890.58])

# Normalize years to start from zero
t = years - years[0]

# Fit the Bass model to iRobot revenue data
p_est, q_est, M_est = fit_bass_model(t, revenue)
print(f"Estimated Parameters: p={p_est:.4f}, q={q_est:.4f}, M={M_est:.2f}")

# Predict revenue using the fitted Bass model
years_future_irobot = np.arange(2012, 2033)  # Extending the prediction for iRobot
predicted_revenue_irobot = predict_bass_model(years_future_irobot, p_est, q_est, M_est)

# Compute cumulative adoption and instantaneous adoption rate for iRobot
cumulative_adoption_irobot = compute_cumulative_adoption(predicted_revenue_irobot, M_est)
instantaneous_adoption_rate_irobot = compute_instantaneous_adoption(p_est, q_est, cumulative_adoption_irobot)

# Plot iRobot revenue trend
plot_data(years, revenue, "iRobot Revenue Over Time", "Year", "Revenue (Million USD)")
```

```python
# Assume Matic Robots Matic has a larger market size than iRobot
M_matic = M_est * 100  # Scaling market potential for a larger reach

# Predict adoption for Matic Robots Matic using the same p, q, but with a larger M
years_future_matic = np.arange(2024, 2041)
predicted_adoption_matic = predict_bass_model(years_future_matic, p_est, q_est, M_matic)

# Compute cumulative adoption fraction for Matic Robots Matic
cumulative_adoption_matic = compute_cumulative_adoption(predicted_adoption_matic, M_matic)

# Compute instantaneous adoption rate f(t) for Matic Robots Matic
instantaneous_adoption_rate_matic = compute_instantaneous_adoption(p_est, q_est, cumulative_adoption_matic)
```

```python
# Plot predicted adoption for Matic Robots Matic
plot_data(years_future_matic, predicted_adoption_matic, "Predicted Adoption of Matic Robots Matic using Bass Model", 
          "Year", "Estimated Number of Adopters")

# Plot the cumulative adoption curve for Matic Robots Matic
plot_data(years_future_matic, cumulative_adoption_matic, "Cumulative Adoption Prediction for Matic Robots Matic", 
          "Year", "Cumulative Adoption Fraction", linestyle='--', color='red')

# Plot the instantaneous adoption rate f(t) for Matic Robots Matic
plot_data(years_future_matic, instantaneous_adoption_rate_matic, "Instantaneous Adoption Rate Over Time", 
          "Year", "Adoption Rate", color='green')
```
